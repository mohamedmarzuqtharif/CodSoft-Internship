```python
import pandas as pd
import matplotlib.pyplot as plt

# Load COVID-19 dataset
# Make sure to replace 'your_dataset.csv' with the path to your actual dataset.
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
data = pd.read_csv(url)

# Data Processing
# Summarize data by country
country_data = data.groupby('Country/Region').sum().iloc[:, 4:]  # Sum over all dates
country_data = country_data.T  # Transpose to have dates as rows

# Convert index to datetime
country_data.index = pd.to_datetime(country_data.index)

# Select a specific country for visualization
country = 'China'  # Change this to visualize other countries
china_data = country_data[country]

# Visualization
plt.figure(figsize=(12, 6))
plt.plot(china_data.index, china_data.values, label='Confirmed Cases', color='blue')
plt.title(f'COVID-19 Confirmed Cases in {country}')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: Cricket player statistics by format
data = {
    'Player': ['Virat Kohli', 'Babar Azam', 'Steve Smith', 'Kane Williamson', 'Joe Root',
               'Jasprit Bumrah', 'Mitchell Starc', 'Kagiso Rabada', 'Brett Lee', 'Shane Warne'],
    'T20_Runs': [3000, 2800, 2000, 1500, 1800, 500, 300, 200, 400, 100],
    'ODI_Runs': [12809, 13043, 8779, 7563, 8856, 1200, 600, 400, 1500, 3000],
    'Test_Runs': [4000, 3000, 5000, 6000, 7000, 0, 0, 0, 0, 3000],
    'T20_Wickets': [50, 30, 20, 10, 15, 100, 90, 80, 20, 0],
    'ODI_Wickets': [0, 0, 0, 0, 0, 120, 100, 80, 300, 708],
    'Test_Wickets': [0, 0, 0, 0, 0, 0, 50, 40, 0, 708],
    'T20_Matches': [100, 90, 80, 70, 60, 50, 40, 30, 20, 10],
    'ODI_Matches': [250, 120, 130, 140, 150, 70, 50, 40, 70, 140],
    'Test_Matches': [100, 50, 70, 80, 90, 0, 50, 40, 0, 145]
}

# Create DataFrame
df = pd.DataFrame(data)

# Set player names as the index
df.set_index('Player', inplace=True)

# Separate dataframes for each format
t20_stats = df[['T20_Runs', 'T20_Wickets', 'T20_Matches']]
odi_stats = df[['ODI_Runs', 'ODI_Wickets', 'ODI_Matches']]
test_stats = df[['Test_Runs', 'Test_Wickets', 'Test_Matches']]

# Plotting Runs and Wickets for T20
plt.figure(figsize=(18, 18))

# T20 Stats
plt.subplot(3, 1, 1)
t20_stats[['T20_Runs', 'T20_Wickets']].plot(kind='bar', stacked=True)
plt.title('T20 Statistics by Player')
plt.xlabel('Players')
plt.ylabel('Runs / Wickets')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Stats')

# ODI Stats
plt.subplot(3, 1, 2)
odi_stats[['ODI_Runs', 'ODI_Wickets']].plot(kind='bar', stacked=True)
plt.title('ODI Statistics by Player')
plt.xlabel('Players')
plt.ylabel('Runs / Wickets')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Stats')

# Test Stats
plt.subplot(3, 1, 3)
test_stats[['Test_Runs', 'Test_Wickets']].plot(kind='bar', stacked=True)
plt.title('Test Statistics by Player')
plt.xlabel('Players')
plt.ylabel('Runs / Wickets')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Stats')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()

# Display the separate dataframes
print("T20 Statistics:")
print(t20_stats)

print("\nODI Statistics:")
print(odi_stats)

print("\nTest Statistics:")
print(test_stats)
```
